2. From the Pillow module, how do you get the RGBA value of any images?


```python
using getcolour can get the RGBA value of image.
```

```python
# importing Image module from PIL package
from PIL import Image, ImageColor

# using getcolor
#L -single “gray-scale” integer value represent the brightness of each pixel
im = ImageColor.getcolor("orange", "L")
print(im)

im1 = ImageColor.getcolor("red", "L")
print(im1)

im2 = ImageColor.getcolor("Blue", "L")
print(im2)
```
